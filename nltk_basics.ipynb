```python
# *finding how many times a word occurs in a text and percentage of that word in text*

print("No of lols in text5:",text5.count('lol'))

print("% of lol:",100*text5.count('lol')/len(text5))
```

```python
# defining helpful functions

def lexical_diversity(text):
    '''computes lexical richness/diversity of a text'''
    return len(set(text))/len(text)

def percentage(word,text):
    '''computes % composition of word/token in text'''
    return text.count(word)/len(text)
```

```python
lexical_diversity(text1)
```

```python
percentage('the',text1)
```

## Selection of relevant words

selecting long words to add more meaning to the context of text , selecting an eqation from set theory :

`	{w | w  V & P(w)}  `

where p is a function of selecting long words (len>10)

```python
def p(w):
    if len(w)>10:
        return True
    return False
```

```python
v=set(text1)

[w for w in v if p(w)] 
```

```python
# since most of the long words could be unique we can find most occuring long words to add more meaning to context

p0=lambda w : True if fdist[w]>5 else False

[w for w in v if p(w) and p0(w)] 
```
